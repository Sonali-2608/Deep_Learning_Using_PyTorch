```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
```

```python
w=torch.tensor(1.0, requires_grad=True)
b=torch.tensor(0.0, requires_grad=True)
```

```python
w
```

```python
b
```

```python
z=w*x+b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn as nn
binary_cross_entropy_loss = nn.BCELoss()
loss = binary_cross_entropy_loss(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
